# ðŸ“ˆ Linear Regression from Scratch â€” Math Behind the Code

## âœ… Goal:

Find the best-fit line:

$$
\hat{y} = mx + b
$$

Where:
- \( m \): **Slope**
- \( b \): **Intercept**
- \( \hat{y} \): Predicted value

---

## ðŸ§® Step 1: Mean of X and Y

$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i \quad , \quad \bar{y} = \frac{1}{n} \sum_{i=1}^{n} y_i
$$

---

## ðŸ§  Step 2: Calculate the Slope \( m \)

$$
m = \frac{
\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})
}{
\sum_{i=1}^{n} (x_i - \bar{x})^2
}
$$

This is also known as:

$$
m = \frac{\text{Cov}(X, Y)}{\text{Var}(X)}
$$

---

## ðŸ§¾ Step 3: Calculate the Intercept \( b \)

$$
b = \bar{y} - m \cdot \bar{x}
$$

This ensures the line passes through the centroid \( (\bar{x}, \bar{y}) \).

---

## ðŸ”® Step 4: Make Predictions

$$
\hat{y} = mx + b
$$

---

## ðŸ“Š Visualization

- **Red dots**: Actual data points
- **Blue line**: Predicted regression line using \( \hat{y} = mx + b \)

---

## âœ… Summary Table

| Term           | Formula                                                                      |
|----------------|------------------------------------------------------------------------------|
| Mean           | ($$ \bar{x} = \frac{1}{n} \sum x_i $$                                        |
| Slope (m)      | $$ m = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2} $$ |
| Intercept (b)  | $$ b = \bar{y} - m \cdot \bar{x} $$                                          |
| Prediction     | $$ \hat{y} = mx + b $$                                                       |

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 5, 6])

# Mean values
mean_x = np.mean(X)
mean_y = np.mean(y)

# Calculating coefficients (slope and intercept)
numerator = np.sum((X - mean_x) * (y - mean_y))
denominator = np.sum((X - mean_x) ** 2)

slope = numerator / denominator
intercept = mean_y - slope * mean_x

print(f"Calculated coefficients:\nSlope = {slope:.4f}, Intercept = {intercept:.4f}")

# Predict values
y_pred = slope * X + intercept

# Plotting
plt.scatter(X, y, color="red", label="Actual points")
plt.plot(X, y_pred, color="blue", label="Regression line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression from Scratch")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # sklearn expects 2D array
y = np.array([3, 4, 2, 5, 6])

# Initialize and fit the model
model = LinearRegression()
model.fit(X, y)

# Get coefficients
slope = model.coef_[0]
intercept = model.intercept_

print(f"Sklearn coefficients:\nSlope = {slope:.4f}, Intercept = {intercept:.4f}")

# Predict
y_pred = model.predict(X)

# Plotting
plt.scatter(X, y, color="red", label="Actual points")
plt.plot(X, y_pred, color="green", label="Sklearn Regression line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression using scikit-learn")
plt.grid(True)
plt.show()
```
