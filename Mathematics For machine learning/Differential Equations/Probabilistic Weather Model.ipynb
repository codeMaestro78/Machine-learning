```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=np.array([[0.5,0.5,0.25],[0.25,0.0,0.25],[0.25,0.5,0.5]])
print(A)
```

```python
xtoday=np.array([[1],[0],[0]])
print(xtoday)
```

```python
print(A@xtoday)
```

```python
for k in range(1,50):
    xtommorow=A@xtoday
    print(f"<---{k}--->\n", xtommorow)
    xtoday=xtommorow
```
