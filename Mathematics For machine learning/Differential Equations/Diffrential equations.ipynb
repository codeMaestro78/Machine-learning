```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
### linear ode
def linear_ode(x,y):
    return np.sin(x)-2*y

solve_linear=solve_ivp(linear_ode,[0,10],[1],dense_output=True)

x_vals=np.linspace(0,10,200)
y_linear=solve_linear.sol(x_vals)[0]
```

```python
### non linear ode
r,K=1,10
def nonlinear_ode(x,y):
    return r*y*(1-(y/K))

sol_nonlinear = solve_ivp(nonlinear_ode, [0, 10], [1], dense_output=True)
y_nonlinear=sol_nonlinear.sol(x_vals)[0]
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x_vals, y_linear, label="Linear ODE")
plt.title("Linear ODE Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.plot(x_vals, y_nonlinear, 'r', label="Nonlinear ODE (Logistic)")
plt.title("Nonlinear ODE Solution")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
