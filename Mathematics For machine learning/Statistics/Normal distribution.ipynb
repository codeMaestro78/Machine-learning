# The Normal Distribution (Gaussian Distribution)

The **Normal Distribution**, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric about its mean. It is the most important probability distribution in statistics because it accurately describes the distribution of many natural phenomena and is fundamental to many statistical methods. Its graph is a bell-shaped curve. ðŸ””



---

## Key Formulas

The behavior and shape of the normal distribution are defined by two parameters: the **mean** ($\mu$) and the **standard deviation** ($\sigma$).

### Probability Density Function (PDF)

The **Probability Density Function (PDF)** describes the likelihood of a random variable taking on a specific value. For the normal distribution, the PDF formula is:

$$ f(x | \mu, \sigma^2) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} $$

Where:
* `x` = the value of the random variable
* `Î¼` (mu) = the **mean** of the distribution (the center of the peak)
* `Ïƒ` (sigma) = the **standard deviation** of the distribution (how spread out the data is)
* `Ï€` (pi) = the constant ~3.14159...
* `e` = Euler's number, the base of the natural logarithm, ~2.71828...

The height of the curve for a given value `x` is determined by this function.

---

### The Standard Normal Distribution

A special case of the normal distribution is the **Standard Normal Distribution**, where the mean is 0 and the standard deviation is 1.

* **Mean ($\mu$)**: 0
* **Standard Deviation ($\sigma$)**: 1

The PDF simplifies to:
$$ f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} $$

### Z-Score Formula

To work with any normal distribution, we can standardize it by converting values into **Z-scores**. A Z-score measures how many standard deviations a data point `x` is from the mean `Î¼`.

The formula for the Z-score is:

$$ Z = \frac{x - \mu}{\sigma} $$

This transformation allows us to use a standard table (Z-table) to find probabilities for *any* normally distributed variable.

---

## Key Characteristics

* **Symmetry**: The distribution is perfectly symmetrical around its mean, `Î¼`.
* **Mean, Median, and Mode**: The mean, median, and mode are all equal and located at the center of the distribution.
* **The Empirical Rule (68-95-99.7 Rule)**: This rule describes the percentage of data that falls within a certain number of standard deviations from the mean.
    * Approximately **68%** of the data falls within 1 standard deviation of the mean (`Î¼ Â± 1Ïƒ`).
    * Approximately **95%** of the data falls within 2 standard deviations of the mean (`Î¼ Â± 2Ïƒ`).
    * Approximately **99.7%** of the data falls within 3 standard deviations of the mean (`Î¼ Â± 3Ïƒ`).
* **Area Under the Curve**: The total area under the curve is equal to 1 (or 100%), representing the total probability of all possible outcomes.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import sigma
```

```python
def normal_distribution(x,mu, sigma):
    prefactor=1 / (sigma*np.sqrt(2 * np.pi))
    return prefactor * np.exp((-1/2)*((x-mu)/sigma)**2)


x_values=np.linspace(0,5,30)

# plt.plot(x_values,y_values,linewidth=2,label='Normal distribution')
plt.plot(x_values,normal_distribution(x_values,0,1),label='Î¼=0, Ïƒ=1', linewidth=2)
plt.plot(x_values,normal_distribution(x_values,0,2),label='Î¼=0, Ïƒ=2', linewidth=2)
plt.plot(x_values,normal_distribution(x_values,0,0.5),label='Î¼=0, Ïƒ=0.5', linewidth=2)
plt.plot(x_values,normal_distribution(x_values,2,1),label='Î¼=0, Ïƒ=1', linewidth=2)

plt.fill_between(x_values,normal_distribution(x_values,0,1),alpha=0.5,color='blue',label='area under curve')
plt.title('Normal Distribution (Î¼=0, Ïƒ=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axvline(mu, color='red', linestyle='--', label='Mean (Î¼)')
plt.legend()
plt.tight_layout()
plt.show()
```
