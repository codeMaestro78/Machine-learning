```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial,exp
```

```python
#  Problem Setup:
# A large company has 10,000 users.
#
# Each user independently sends a request to the server with a very small probability in a given second (say,
# ùëù = 0.0005
# p=0.0005).
#
# You want to model the number of requests arriving at the server in one second.
```

```python
n = 10000         # number of users trials
p = 0.0005        # probability of sending a request
lambda_poisson = n * p  # average expected requests per second

# range of events to calculate the prob for 0 to 20
x=np.arange(0,21)
```

Binomial vs Poisson Distribution ‚Äî Manual Explanation



üîß Scenario Setup

We assume:





Number of users (n): 10,000



Probability per user (p): 0.0005



Expected rate (Œª):

$$ \lambda = n \cdot p = 5 $$

We want to model the number of events (e.g., requests arriving at a server in one second).



üìò Binomial Distribution

The Binomial Probability Mass Function (PMF) is:

$$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $$

Where:





$n$ = number of trials



$k$ = number of successes



$p$ = probability of success on a single trial



üìó Poisson Distribution

The Poisson Probability Mass Function (PMF) is:

$$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} $$

Where:

$\lambda$ = expected number of events ($n \cdot p$)

$k$ = number of actual events

$e$ = Euler‚Äôs number ($\approx 2.718$)

üìå Key Relationship

When $n$ is large and $p$ is small, the Binomial distribution can be approximated by the Poisson distribution with:

$$ \lambda = n \cdot p $$

```python
# Binomial PMF
def binomial_pmf(k,n,p):
    comb=factorial(n)//(factorial(k)* factorial(n-k))
    return comb*(p**k)*((1-p)**(n-k))

binomial_probs=np.array([binomial_pmf(k,n,p) for k in x])
binomial_probs
```

```python
# Poisson PMF
def poisson_pmf(k,lam):
    return (lam**k)*exp(-lam)/ factorial(k)

poisson_probs=np.array([poisson_pmf(k,lambda_poisson) for k in x ])
poisson_probs
```

```python
# visualization

plt.figure(figsize=(12,6))
plt.plot(x,binomial_probs,'o--',label='binomial pmf',color='blue')
plt.plot(x, poisson_probs, 's--', label='Poisson pmf)', color='orange')
plt.title("Binomial vs Poisson (Manual Calculation)")
plt.xlabel("Number of Requests in One Second (k)")
plt.ylabel("Probability P(X = k)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
