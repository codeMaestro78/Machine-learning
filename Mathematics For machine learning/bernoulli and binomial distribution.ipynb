# Bernoulli and Binomial Random Variables

## 1. Bernoulli Random Variable

### Definition
A **Bernoulli random variable** takes values:
- 1 (success) with probability $p$
- 0 (failure) with probability $1 - p$

Represents a single trial outcome.

### Probability Mass Function (PMF)
$P(X = x) = p^x (1 - p)^{1 - x}$, for $x \in \{0, 1\}$

### Mean
$E[X] = p$

### Variance
$Var(X) = p(1 - p)$

### Example
Visitor clicks a signup button with $p = 0.2$:
- $P(X = 1) = 0.2$, $P(X = 0) = 0.8$

---

## 2. Binomial Random Variable

### Definition
Counts successes in $n$ independent Bernoulli trials, each with probability $p$.

### Probability Mass Function (PMF)
$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$, for $k = 0, 1, \dots, n$

### Mean
$E[X] = np$

### Variance
$Var(X) = np(1 - p)$

### Example
Email sent to 10 people, each with 30% reply chance. Probability of exactly 3 replies:
$P(X = 3) = \binom{10}{3} (0.3)^3 (0.7)^7 \approx 0.2668$

---

## Summary Table

| Feature   | Bernoulli           | Binomial                     |
|-----------|---------------------|------------------------------|
| Trials    | 1                   | $n$                          |
| Outcomes  | 0 or 1              | 0, 1, ..., $n$               |
| Mean      | $p$                 | $np$                         |
| Variance  | $p(1 - p)$          | $np(1 - p)$                  |
| Formula   | $p^x (1 - p)^{1 - x}$ | $\binom{n}{k} p^k (1 - p)^{n - k}$ |

---

## Visualization
Bar plots for:
- Bernoulli distribution ($p = 0.2$)
- Binomial distribution ($n = 10$, $p = 0.3$)

![Bernoulli and Binomial Distributions](bernoulli_binomial.png)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli, binom

# Parameters
p_bernoulli = 0.2
n_binomial = 10
p_binomial = 0.3

# Bernoulli PMF
x_bernoulli = [0, 1]
pmf_bernoulli = bernoulli.pmf(x_bernoulli, p_bernoulli)

# Binomial PMF
x_binomial = np.arange(0, n_binomial + 1)
pmf_binomial = binom.pmf(x_binomial, n_binomial, p_binomial)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Bernoulli Plot
axs[0].bar(x_bernoulli, pmf_bernoulli, color='skyblue')
axs[0].set_title("Bernoulli Distribution (p=0.2)")
axs[0].set_xlabel("Outcome")
axs[0].set_ylabel("Probability")
axs[0].set_xticks([0, 1])

# Binomial Plot
axs[1].bar(x_binomial, pmf_binomial, color='salmon')
axs[1].set_title("Binomial Distribution (n=10, p=0.3)")
axs[1].set_xlabel("Number of Successes")
axs[1].set_ylabel("Probability")

plt.tight_layout()
plt.show()
```
