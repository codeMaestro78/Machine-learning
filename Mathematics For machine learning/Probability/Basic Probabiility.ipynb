## 1. Basic Probability
This is the foundation for calculating the likelihood of a single event.

### ðŸ”¹ Formula
The probability of an event *A* is the ratio of favorable outcomes to the total possible outcomes.
$$P(A) = \frac{\text{Number of Favorable Outcomes}}{\text{Total Number of Outcomes}}$$

### âœ… When to Use
Use this rule when all outcomes are **equally likely**.
* **Examples:** Tossing a fair coin, rolling a standard die, or drawing a card from a well-shuffled deck.

***

## 2. Complement Rule
This rule helps find the probability that an event will *not* occur.

### ðŸ”¹ Formula
The probability of the complement of event *A* (denoted $A^c$) is 1 minus the probability of *A*.
$$P(A^c) = 1 - P(A)$$

### âœ… When to Use
Use this when it's simpler to calculate the probability of the event **happening** and then subtract it from 1.
* **Example:** What is the probability of *not* rolling a 6 on a die? It's easier to find the probability of rolling a 6 ($1/6$) and calculate $1 - 1/6 = 5/6$.

***

## 3. Addition Rule
This rule is for finding the probability of at least one of two events occurring.

### ðŸ”¹ Formula
The probability of event *A* or event *B* occurring is the sum of their individual probabilities minus the probability of both occurring together.
$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$
*Note: If events A and B are **mutually exclusive** (cannot happen at the same time), then $P(A \cap B) = 0$, and the formula simplifies to $P(A \cup B) = P(A) + P(B)$.*

### âœ… When to Use
Use this to find the probability of **either A or B** (or both) happening.
* **Example:** What is the probability of drawing a King or a Heart from a deck of cards?

***

## 4. Conditional Probability
This measures the probability of an event occurring, given that another event has already occurred.

### ðŸ”¹ Formula
The probability of event *A* given that event *B* has occurred is:
$$P(A \mid B) = \frac{P(A \cap B)}{P(B)}$$

### âœ… When to Use
Use this when the outcome of one event **affects the probability** of another.
* **Example:** What is the probability that a student passed a test (*A*), given that they studied (*B*)?

***

## 5. Bayes' Theorem
This theorem describes the probability of an event based on prior knowledge of conditions that might be related to the event. It's a way to "reverse" conditional probability.

### ðŸ”¹ Formula
$$P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}$$

### âœ… When to Use
Use this for updating probabilities based on new evidence.
* **Applications:** Medical diagnoses (e.g., probability of a disease given a positive test), spam filtering, and machine learning models.

***

## 6. Multiplication Rule (for Intersections)
This rule is used to find the probability that two (or more) events will occur together. The formula depends on whether the events are independent.

### ðŸ”¹ Formulas
* **For Independent Events:** If events *A* and *B* do not influence each other.
    $$
    P(A \cap B) = P(A) \cdot P(B)
    $$
* **For Dependent Events:** If the outcome of event *A* affects the outcome of event *B*.
    $$
    P(A \cap B) = P(A) \cdot P(B \mid A)
    $$

### âœ… When to Use
Use this to find the probability of **both A and B** happening.
* **Independent Example:** The probability of flipping two heads in a row.
* **Dependent Example:** The probability of drawing two aces from a deck without replacement.

***

## 7. Permutations and Combinations
These are used for counting outcomes, which is essential for calculating basic probabilities.

### ðŸ”¹ Formulas
* **Permutations (Order Matters):** The number of ways to arrange *r* items from a set of *n* items.
    $$
    _n P_r = \frac{n!}{(n - r)!}
    $$
* **Combinations (Order Doesn't Matter):** The number of ways to choose *r* items from a set of *n* items.
    $$
    _n C_r = \binom{n}{r} = \frac{n!}{r!(n - r)!}
    $$

### âœ… When to Use
* Use **permutations** when the sequence is important (e.g., arranging letters in a word, finishing places in a race).
* Use **combinations** when the sequence is irrelevant (e.g., picking a committee of 3 people from a group of 10, choosing lottery numbers).
